#Exercício 02) Faça um programa que permita ao usuário digitar o seu nome e em seguida o mostre o nome do usuário de trás para frente utilizando somente letras maísculas.

```python
nome = input("Digite o seu nome:")
```

```python
nome_invertido = nome[::-1]
```

```python
nome_maiusculo = nome_invertido.upper()
```

```python
print(nome_maiusculo)
```

#3) Faça um programa que permita o usuário digitar uma frase e:

```python
frase = input('Digite uma frase:')
```

```python
print(frase)
```

##Imprima a frase, desprezando os quatro últimos dígitos:

```python
print(frase[:-4])
```

##Imprima apenas a primeira letra em maiúsculo:

```python
print(frase.capitalize())
```

##Imprima cada letra da frase em maiúsculo:

```python
print(frase.title())
```

##Imprima a frase, substituindo " " por "_":

```python
print(frase.replace(" ", "_"))
```

#5) Faça um programa que peça o raio de um círculo, calcule e mostre sua área.

```python
raio = float(input("Digite o raio do círculo: "))
```

##Calcule e mostre sua área:

```python
pi = 3.14159
area = pi * (raio**2)
```

```python
print(area)
```
