#10) Faça um Programa que leia 20 números inteiros e armazene-os num vetor. Armazene os números pares no vetor PAR e os números IMPARES no vetor ímpar. Imprima os três vetores.

```python
vetor = []
vetor_par = []
vetor_impar = []
for i in range(1,20):
    num = int(input(f"Digite o número {i}: "))
    vetor.append(num)
    if num % 2 == 0: #se o resto da divisão for igual a zero
        vetor_par.append(num) 
    else:
        vetor_impar.append(num) 

print(vetor)
print(f"O vetor par é: {vetor_par}")
print(f"O vetor ímpar é: {vetor_impar}")
```

#11) Foram anotadas as idades e alturas de 5 alunos. Faça um Programa que liste quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos.

## Resolução com dois vetores:

```python
lista_idade = []
lista_altura = []
soma = 0.

for i in range(1,5):
    idade =  int(input(f"Digite a idade do aluno {i}: "))
    lista_idade.append(idade)
    altura =  float(input(f"Digite a altura do aluno {i}: "))
    lista_altura.append(altura)
    soma += altura
    print('_'*20)

media_altura = soma / len(lista_altura)

print(f"A média de altura é: {media_altura} ")

length_lista = len(lista_idade)
   
# Iterar a lista pelo índice
for i in range(length_lista):
    if lista_idade[i] > 13 and lista_altura[i] < media_altura:
       print(f"O aluno {i+1} possui altura inferior: {lista_altura[i+1]}. ") 
```

##Resolução com matriz

```python
alunos = []

soma = 0.

for i in range(0,5):
    nome = input(f"Digite o nome do aluno {i}: ")
    idade =  int(input(f"Digite a idade do aluno {i}: "))
    altura =  float(input(f"Digite a altura do aluno {i}: "))
    alunos.append([nome, idade, altura])
    soma += altura
    print('_'*20)

media_altura = soma / len(alunos)

print(f"A média de altura é: {media_altura} ") 
```

```python
print(alunos)
```

```python
# Iterar a matriz pelo conteúdo
for aluno in alunos:
    if aluno[1] > 13 and aluno[2] < media_altura:
       print(f"O(a) aluno(a) {aluno[0]} cuja idade é {aluno[1]} possui altura inferior: {aluno[2]}. ") 
```

#12) Faça um programa que peça ao usuário o primeiro elemento de uma P.A, a razão da P.A e o número de elementos a serem exibidos.

Fórmula: an = a1 + (n - 1) * razão

```python
a1 = int(input("Digite o número inicial da PA: "))
```

```python
razao = int(input("Digite a razão da PA: "))
```

```python
termos = int(input("Digite o número de termos da PA: "))
```

```python
count = 1
while(count <= termos):
  num = a1 + (count - 1) * razao
  print(f"Termo {count}: {num}")
  count += 1
```

##13) Faça uma função que percorra uma lista de 10 palavras e retorne "Sim" ou "Não" caso tenha encontrado o termo na lista.

```python
lista = []
for i in range(1,11):
  palavra = input('Digite uma palavra: ')
  lista.append(palavra)
```

```python
print(lista)
```

```python
def busca(lista_palavras, termo):
  retorno = "Não"
  for palavra in lista_palavras:
    print(palavra)
    if palavra == termo:
      retorno = "Sim"
      break
  return retorno
```

```python
busca(lista, "uva")
```
