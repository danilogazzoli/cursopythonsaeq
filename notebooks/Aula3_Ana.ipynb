Aula 3 - Estatística 

```python
# Importação dos pacotes

import numpy
import statistics
```

Resumo das funções:

Biblioteca: **numpy**

- média > mean()
- mediana > median()
- variância amostral > var()
- desvio padrão amostral > std()

Biblioteca: **statistics**

- moda > mode()


```python
# Média aritmética simples

dados = [99,86,87,88,111,86,103,87,94,78,77,85,86]

media = numpy.mean(dados)

print(media)
```

```python
# Mediana

# Mediana com n ímpar (13)

dados = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mediana_impar = numpy.median(dados)

print(mediana_impar)
```

```python
# Mediana com n par (12)

dados = [99,86,87,88,86,103,87,94,78,77,85,86]

mediana_par = numpy.median(dados)

print(mediana_par)
```

```python
# Moda com o pacote statistics 

dados = [99,86,87,88,111,86,103,87,94,78,77,85,86]

moda = statistics.mode(dados)     

print(moda)
```

```python
# Amplitude

dados = [32,111,138,28,59,77,97]

amplitude = (numpy.max(dados)) - (numpy.min(dados))

print(amplitude)
```

```python
# Variância 

dados = [32,111,138,28,59,77,97]

variancia = numpy.var(dados)

print(variancia)
```

```python
# Desvio Padrão

dados = [32,111,138,28,59,77,97]

desvio_padrao = numpy.std(dados)

print(desvio_padrao)
```
